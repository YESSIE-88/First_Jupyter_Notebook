# Importing the pyplot library to make a bar chart

```python
import matplotlib.pyplot as plt
```

Defining some arbitrary values

```python
x = [10, 20, 30, 40, 50, 60]
y = [8, 16, 32, 64, 16, 32]
```

Defining our chart

```python
plt.title('First Bar Graph')
plt.bar(x, y, color='red', width=5)
plt.show()
```

# Importing the seaborn library to make a line plot

```python
import seaborn as sns
```

Using a predefined dataset

```python
data = sns.load_dataset("iris")
```

Ploting our data

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```
